# Khám phá dữ liệu mẫu

Đọc ba file dữ liệu mẫu và khám phá tổng quan về cấu trúc cùng nội dung chính của từng bảng.

```python
from pathlib import Path
import pandas as pd

base_path = Path(r"D:\retail_lakehouse\data\raw")

customer_path = base_path / "customer" / "customers_sample.csv"
product_path = base_path / "product" / "products_sample.csv"
sales_path = base_path / "sales" / "sales_sample.csv"

customers = pd.read_csv(customer_path)
products = pd.read_csv(product_path)
sales = pd.read_csv(sales_path)

customers.head(), products.head(), sales.head()
```

```python
print("Customers shape:", customers.shape)
print("Products shape:", products.shape)
print("Sales shape:", sales.shape)

print('\nCustomers columns:')
print(customers.columns.tolist())
print('\nProducts columns:')
print(products.columns.tolist())
print('\nSales columns:')
print(sales.columns.tolist())
```

```python
for name, df in [("Customers", customers), ("Products", products), ("Sales", sales)]:
    print(f"\n=== {name} overview ===")
    display(df.info())
    print("Missing values by column:")
    print(df.isna().sum())
    print("\nSample value counts for each column:")
    for col in df.columns:
        sample_values = df[col].dropna().unique()[:8]
        print(f"  - {col}: {len(df[col].unique())} unique values; sample -> {sample_values}")
```
